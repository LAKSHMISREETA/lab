```python
# Define the Car class
class Car:
    # Constructor to initialize the attributes
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.started = False
    
    # Method to start the car
    def start(self):
        self.started = True
        print(f"The {self.year} {self.make} {self.model} is now started.")
    
    # Method to stop the car
    def stop(self):
        self.started = False
        print(f"The {self.year} {self.make} {self.model} is now stopped.")
    
    # Method to display car details
    def display_details(self):
        print(f"Car Details: {self.year} {self.make} {self.model}")
        print("Status: " + ("Started" if self.started else "Stopped"))

# Creating instances (objects) of the Car class
car1 = Car("Toyota", "Corolla", 2021)
car2 = Car("Ford", "Mustang", 2020)

# Using the methods of the instances
car1.display_details()
car1.start()
car1.display_details()

car2.display_details()
car2.start()
car2.stop()
car2.display_details()
```
